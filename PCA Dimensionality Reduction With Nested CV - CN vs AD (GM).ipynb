```python
inputDataFrame = pd.read_csv("D:\\Fresh Alzheimers Work\\Dataset\\CN vs AD\\GM\\train_data_set_GM_CN_vs_AD.csv")
inputDataFrame.shape
```

```python
X_data = inputDataFrame.drop(columns = ['subject_Id','Group'])
```

```python
X_data = X_data.values
```

```python
Y_target = inputDataFrame['Group']
```

```python
Y_target = Y_target.values
```

```python
Y_target.shape
```

```python
svm_pipe = Pipeline(steps=[('scaling', StandardScaler()), ('dimensionality_reduction',PCA()),('classifier', SVC())])
```

```python
number_of_components = list(range(1,10))
```

```python
svm_C = [0.1,0.01]
svm_gamma = [0.1,0.01]
svm_kernel = ['linear']
```

```python
svm_param_grid = dict(dimensionality_reduction__n_components = number_of_components, classifier__C = svm_C, classifier__gamma = svm_gamma,classifier__kernel = svm_kernel)
```

```python
svm_estimator = GridSearchCV(svm_pipe,svm_param_grid,n_jobs=30,scoring='accuracy',cv = RepeatedStratifiedKFold(n_splits = 10, n_repeats = 10, random_state = 2), verbose = 20)
```

```python
svm_outer_cv = StratifiedKFold(n_splits=10, random_state=1)
```

```python
svm_cv_results_list = []
svm_best_params_list = []
svm_best_params_accuracy_list = []
```

```python
count = 0
for train_index,test_index in svm_outer_cv.split(X_data,Y_target):
    count = count + 1
    print("outer cv iteration number: "+str(count))
    
    X_train_outer, X_test_outer = X_data[train_index], X_data[test_index]
    Y_train_outer, Y_test_outer = Y_target[train_index], Y_target[test_index]
  
    print("************* begin fitting inner GridSearchCV with RepeatedStartifiedKFold ***************")
    svm_estimator.fit(X_train_outer,Y_train_outer)
    print("*************  end fitting inner GridSearchCV with RepeatedStartifiedKFold ***************")
    
    print("****** begin predicting outer cv test data *******")
    Y_predicted = svm_estimator.predict(X_test_outer)
    score = accuracy_score(y_true=Y_test_outer,y_pred=Y_predicted)
    print("accuracy: "+str(score))
    print("****** end predicting outer cv test data *******")
    
    svm_best_params_accuracy_list.append(score)
    svm_cv_results_list.append(svm_estimator.cv_results_)
    svm_best_params_list.append(svm_estimator.best_params_)
    
```
