```python
import pandas as pd
```

```python
from sklearn.decomposition import PCA
```

```python
inputDataFrame = pd.read_csv("E:\\Fresh Alzheimers - Test run\\Train data set\\feature_vector_CN_vs_AD.csv")
```

```python
inputData = inputDataFrame.drop(columns=['file','Group'])
```

```python
inputData.head()
```

```python
pca = PCA(0.99)
```

```python
pca.fit(inputData)
```

```python
transformedTrainData = pca.transform(inputData)
```

```python
tempTrainDf = pd.DataFrame(data = transformedTrainData)
```

```python
tempGroupDf = pd.DataFrame(data = inputDataFrame[['file','Group']])
```

```python
newdf = pd.concat([tempTrainDf,tempGroupDf],axis=1)
```

```python
from sklearn.utils import shuffle
```

```python
shuffledDf = shuffle(newdf)
```

```python
newdf
```

```python
shuffledDf
```

```python
shuffledDf.to_csv("C:\\Fresh Alzheimers Work\\Data sets\\GM\\CN vs AD\\pca_train_data_set_CN_vs_AD.csv",index=False)
```

```python
testDataFrame = pd.read_csv("E:\\Fresh Alzheimers - Test run\\Test data set\\test_data_set_CN_vs_AD.csv")
```

```python
testData = testDataFrame.drop(columns=['file','Group'])
```

```python
transformedTestData = pca.transform(testData)
```

```python
transformedTestData.shape
```

```python
tempTestDf = pd.DataFrame(data = transformedTestData)
testDf = pd.concat([tempTestDf,testDataFrame[['file','Group']]],axis=1)
```

```python
testDf.to_csv("C:\\Fresh Alzheimers Work\\Data sets\\GM\\CN vs AD\\Test\\pca_test_data_set_CN_vs_AD.csv",index=False)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticRegr = LogisticRegression(solver = 'lbfgs')
```
